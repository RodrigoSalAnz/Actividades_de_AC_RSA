# **Ejercicio 1**: Aproximación de π

Una forma para estimar el valor de π (3.141592...) es usando el método de Monte Carlo. Primero,
tomamos un cuadrado de 1 × 1 y un círculo inscrito de radio 1/2, generamos una cantidad arbitraria
de puntos uniformemente distribuidos sobre la superficie del cuadrado y coloreamos de rojo aquellos
que se encuentren sobre la superficie del círculo, y de azul, aquellos que estén fuera.

**1. Escribe un algoritmo que estime el valor de π y que te permita visualizar algo similar al gráfico
de la Figura 1, asegúrate de incluir el conteo del número de puntos rojos, número de puntos
totales, y la respectiva estimación de π.**

**Solución:**

```julia
n=5000 #Este es el número de puntos que quiero que dibuje.

#Se usarán arrays para almacenar las coordenadas de los puntos.

A=[] #Aquí se guardaran las abcisas de las coordenadas de los puntos.
B=[] #Aquí se guardaran las ordenadas de las coordenadas de los puntos.
ROJ_A=[] #Aquí se guardaran las abcisas de las coordenadas de los puntos que se encuentran dentro del círculo.
ROJ_B=[] #Aquí se guardaran las ordenadas de las coordenadas de los puntos que se encuentran dentro del círculo.
AZU_A=[] #Aquí se guardaran las abcisas de las coordenadas de los puntos que se encuentran fuera del círculo.
AZU_B=[] #Aquí se guardaran las ordenadas de las coordenadas de los puntos que se encuentran fuera del círculo.




for i in 1:n #En este ciclo for se se generarán de manera aleatoria las coordenadas de los "n" puntos. 
    A2 = insert!(A, i, rand()) #Aquí se genera un valor aleatorio y es almacenado en el array "A" en el lugar "i".
    B2 = insert!(B, i, rand()) #Aquí se genera un valor aleatorio y es almacenado en el array "B"  en el lugar "i".
end
```

```julia
#Las coordenadas de los puntos están dados por (A[i],B[i]) con 1<=i<=n. Para identificar los puntos que están dentro o fuera 
#del círculo, se utiliza la ecuación de la circunferencia con radio r=0.5 y centro en (0.5,0.5), pero en nuestro caso en vez 
#de usar una ecuación usamos una inecuación para selecionar los puntos que están dentro y en el borde de la circuneferncia.
 
for i in 1:n

    if ((A[i]-1/2)^2+(B[i]-1/2)^2)^(1/2)<=0.5 #Si el punto se encuntra dentro del círculo, sus coordenadas se guardaran en 
                                              #los arrays ROJ_A y ROJ_B respectivamente.
        ROJ_2=push!(ROJ_A, A[i]) #La función push se encarga agregar A[i] hasta el final del array "ROJ_A" (análogo a las demás) 
        ROJ_3=push!(ROJ_B, B[i])
                                              #Si el punto no se encuntra dentro del círculo, sus coordenadas se guardaran en 
                                              #los arrays AZU_A y AZU_B respectivamente.
    else

        AZU_2=push!(AZU_A, A[i])
        AZU_3=push!(AZU_B, B[i])

    end

end
```

```julia
#Estos ciclos for se encargan de contar los elementos de los los arrays para saber cuantos puntos están dentro o fuera del círculo.  

contR=0 #Esta variable almacenará el número puntos que están dentro de la circunferencia.
contA=0 #Esta variable almacenará el número puntos que están fuera de la circunferencia.
for i in ROJ_B
    contR=contR+1
end

for i in AZU_B
    contA=contA+1
end
```

**Resultado**:

```julia
println("Número de puntos totales: ",n)
println("Número de puntos dentro del círculo: ",contR)
println("Valor Aproximado de Pi: ", 4*contR/(contR+contA))
println("Error del Valor Aproximado de Pi:", 4*contR/(contR+contA)-pi)
```

**4. Desarrolla un código que visualice la gráfica de una función junto con la recta tangente a un
punto en su dominio, incluye en tu visualización una leyenda que muestre la pendiente de la
recta tangente. Prueba tu implementación con las funciones descritas en los puntos anteriores.**

Tenemos que la ecuación de la recta tangente es:

$$
y - f(a) = f'(x) \cdot (x-a)
$$

$$
\Rightarrow y = f'(a) \cdot (x-a) + f(a)
$$

Donde: \
f(a): es la función evaluada en el punto. \
f'(a): es la derivada de la función evaluada en el punto.

```julia
function RecTan(x,f,fprima,a) # Definimos nuestra función para calcular la recta tangente como 'RecTan' que depende de 'x'
    # (que se graficará junto con f), de 'f' (la función a evaluar), fprima (que es la derivada de la función calculada
    # con el código para calcular derivadas en un punto que realizamos) y 'a' (el punto a evaluar).
    plot(x,f, label = false) # Pedimos a la función que grafique la función 'f' con respecto a 'x'.
    c = fprima*a-f(a) # Hacemos una variable 'c' que es parte de la expresión de la recta tangente.
    y = fprima*x.-c # Calculamos el valor de la recta tangénte en 'a' con ayuda de c.
    plot!(x,y, label = "La pendiente es $fprima") # Graficamos una función modificadora que graficara la pendiente en el punto 'a'
    # sobre el eje 'x' y nos dara su valor.
end
```

```julia
g=function f(a) 
  cos(a)  
end  # Retomamos la función 'g' que utilizamos en los incisos pasados. 
b = 2 # Tomamos un punto en el dominio de la función.
x = range(0,pi,1000) # Hacemos un rango 'x' en donde correra la función. 
gprima = DerivadaParam(g,b,19957) # Calculamos la derivada de la función en el punto 'b' con nuestra función para calcular 
# derivadas en un punto.
RecTan(x,g,gprima,b) # Calculamos y graficamos la pendiente en 'b' junto con la función.
```

```julia
h=function f(a) 
  log(a)  
end  # Retomamos la función 'log(a)' de los incisos anteriores.
b = 2 # Elegimos un punto en el cual calcularemos la pendiente.
x = range(1,10,1000) # Elegimos un rango de 'x' sobre el cual graficar la función.
hprima = DerivadaParam(h,b,19957) # Calculamos la derivada en el punto.
RecTan(x,h,hprima,b) # Graficamos la pendiente en el punto.
```

## Integración

**1. Escribe la función ‘integrar‘ con las siguientes características:**

`Evalúa la integral de una función en un intervalo dado con cierta precisión.`

`con precisión n.`

**Prueba tu implementación con los siguientes valores: f(x) = cos(x) en el intervalo [0, π], el
resultado debería aproximarse a 0.** 

**Solución:**

De Cálculo II sabemos que: Dada una función $f$ acotada en $[a, b]$, $f$ es integrable si y sólo si existe una sucesión de particiones $(P_n)_{n \in \mathbb{N}}$ de $[a,b]$ tal que $\lim\limits_{n \to \infty}(U(f, P_n)-L(f,P_n))=0$. En este caso  $\lim\limits_{n \to \infty}U(f, P_n)=\int_{a}^{b} f(x)  dx= \lim\limits_{n \to \infty}L(f,P_n)$. Como la suma de Riemann de $f$ cumple que: $L(f, P_n)\leq \sum_{i=1}^{n}f(x_{i})(t_i-t_{i-1}) \leq U(f, P_n)$ con $x_i \in [t_{i-1},t_i]$, entonces por el Teorema del Sandwich: $\lim\limits_{n \to \infty}\sum_{i=1}^{n}f(x_{i})(t_i-t_{i-1})=\int_{a}^{b} f(x)  dx$

Como en este ejercicio trataremos con funciones continuas acotadas en un intervalo, se sigue que son integrables(se demuestra en las clases de Cálculo II) y por ende también existe tal sucesión. Puesto que no podemos hacer que en la computadora $n$ tienda a infinito, vamos a tomar una $n$ la computadora tan grande como pueda para obtener una aproximación a la integral. 

La sucesión de particiones que tomaremos será: $P_n={t_0,...,t_n}$ con $t_i=a+\frac{i(b-a)}{n}$, entonces $t_i-t_{i-1}=\frac{(b-a)}{n}$. Por lo que $\sum_{i=1}^{n}f(t_{i-1})(t_i-t_{i-1})=\sum_{i=1}^{n}f(a+\frac{(i-1)(b-a)}{n})\frac{(b-a)}{n}\approx \int_{a}^{b} f(x)  dx$ 


```julia
g=function f(a) #Aquí defino mi función que quiero calcularle su derivada.
  cos(a)  
end

function integrar( func, a, b, n)
    Suma=0 #Esta variable almacenará el valor de la integral.
    for i in 1:n #Este ciclo sumara cada término de la suma de Riemann.
       Suma=Suma+func(a+(b-a)*((i-1)/n))*(b-a)/n 
    end
    return Suma

end
```

```julia
integrar(g,0,π, 10000)
```
